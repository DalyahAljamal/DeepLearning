````julia
#Loss Functions


function loss(w,x,y,l2)# for one mini batch
   # J=mean(abs2,y-predict(w,x));
    y_pred= predict(w,x);
    diff= abs2.(y-y_pred);
     J=sum(diff)/size(y_pred,2);
    if l2 != 0; J += Float32(l2) * sum(sum(abs2,wi) for wi in w[1:2:end]); end
    return J;
end

function loss_all(w,x,y,l2, data) #for the whole data
 return mean( [loss(w,x,y,l2) for (x,y) in data ]);
end


#Huber Loss
function huber_loss(w, x, y,l2 ; e=0.1)
    ypred=predict(w,x)
    N= size(y,1)
    E=e# ill(e, N)
    
    t = sum(y- ypred);
    flag = t > c
    weights=[w[1], w[3], w[5],w[7], w[9], w[11]]
    if flag
        loss=-1* (0.5 .* t.^ 2)- (e * (0.5 * e - t));
        if l2 != 0; loss += Float32(l2) * sum(sum(abs2,wi) for wi in w[1:2:end]); end
        #loss+= l1* sum(weights,2);
        return loss
    else
        loss=-1* (0.5 .* t.^ 2);
        if l2 != 0; loss += Float32(l2) * sum(sum(abs2,wi) for wi in w[1:2:end]); end
        #loss+= l1* sum(weights,2);
        return loss
    end
end


#Gradient Loss
lossgradient = grad(loss);
````

## Accuracy & Joint loss

````julia
function joint_loss_acc(w, data ) # To be added in main file with predict fun
    threshold=60;
    for (x,y) in data
        y_pred= predict(w,x);
        n=size(y_pred,2);
        l=size(y_pred,1);
        num_co=3;
        diff_all=0;
        pos=0;

        for j in 1:n
            for i in 1:num_co:l
                sum_diff = abs2(y[i,j]-y_pred[i,j]) +abs2(y[i+1,j]-y_pred[i+1,j]) + abs2(y[i+2,j]-y_pred[i+2,j]) ;
                sqrt_sum_diff=sqrt(sum_diff);
                diff_norm= sqrt_sum_diff/(l/3);
                diff_all +=diff_norm; 

                #Checking if within threshold
                if diff_norm<=threshold
                    pos+=1;
                end
            end
        end
        joint_loss= diff_all /(length(data));
        acc= (pos/(length(data))) *100;
        return joint_loss , acc
        
    end
end
````

## CNN Model Predict 

````julia
#Conv 8(5*5) , Max Pool (4*4), Conv 8(5*5), Max Pool (2*2), Conv 8(3*3), 2 FC (1024),droupout, PCA (30), FC(48)
function predict(w,x; pdrop=0.3)
  
    x = pool(relu.(conv4(w[1],x) .+ w[2]); window=2) 
    
    #Second Conv Layer
    x = pool(relu.(conv4(w[3],x) .+ w[4]); window=4) 
    
    #Third Conv Layer
    x = relu.(conv4(w[5],x) .+ w[6])
    
    #FC 1
    x = relu.(w[7]*mat(x) .+ w[8])
    
    #FC 2
    x = relu.(w[9]* x .+ w[10])
    
    #Dropout layer
    x= dropout(x,pdrop);
    
    #PCA layer
    
    #OUTPUT
    return w[11]* x .+ w[12]
    
end
````
